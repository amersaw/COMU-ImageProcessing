```python
def convert_RGB_to_Gray(img):
    res = np.zeros((img.shape[0], img.shape[1]))
    for i in range(res.shape[0]):
        for j in range(res.shape[1]):
                res[i,j] = sum(img[i,j,:])/3
    return res
```
